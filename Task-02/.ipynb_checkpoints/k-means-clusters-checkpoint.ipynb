### Preparing the dataset :

```python
X = data.iloc[:,:-1]
y = data.iloc[:,-1]
print('X shape :', X.shape)
print('y shape :', y.shape)
```

```python
data['Species'].describe()
```

### Computing k using the elbow method :

```python
from sklearn.cluster import KMeans
sse = []
k_possible = []
for i in range(1,10):
    kmodel = KMeans(n_clusters=i)
    kmodel.fit(X)
    k_possible.append(i)
    sse.append(kmodel.inertia_)   
plt.figure(figsize=(8,6))
plt.title('Elbow Method', fontsize=16)
plt.scatter(k_possible, sse, c='purple')
plt.plot(k_possible, sse, c='red')
plt.xlabel('Number of clusters (k)', fontsize=12)
plt.ylabel('Sum of Squared distance', fontsize=12)
plt.show()
```

Hence, from the plot one can say that the optimal k value could be 3.

### Clustering the dataset :

```python
kmeans = KMeans(n_clusters = 3, 
                init='k-means++', 
                random_state=0)
y_kmeans = kmeans.fit_predict(X)
print('Clusters ready with k = 3')
```

### Plotting the clusters with centroids :

```python
plt.figure(figsize=(10,8))
plt.title("K = 3", fontsize=16)
plt.scatter(X.SepalLengthCm[y_kmeans == 0],
            X.SepalWidthCm[y_kmeans == 0],
            s=100, c='darkblue', label='Iris-Setosa')
plt.scatter(X.SepalLengthCm[y_kmeans == 1],
            X.SepalWidthCm[y_kmeans == 1], 
            s=100, c='purple', label='Iris-Versicolor')
plt.scatter(X.SepalLengthCm[y_kmeans == 2],
            X.SepalWidthCm[y_kmeans == 2], 
            s=100, c='red', label='Iris-Virginica')
plt.scatter(kmeans.cluster_centers_[:,0],
            kmeans.cluster_centers_[:,1],
            marker='o',
            s=200,c='orange') 
plt.legend()
plt.show()
```
